```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/placement.csv")
df.head()

```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.fillna(df.mean(numeric_only=True), inplace=True)
```

```python
df.replace({"Yes": 1, "No": 0}, inplace=True)
df = df.infer_objects(copy=False)
```

```python
plt.figure()
sns.histplot(df["CGPA"], kde=True)
plt.title("CGPA Distribution")
plt.show()

```

```python
plt.figure()
sns.boxplot(x="Placement", y="CGPA", data=df)
plt.title("Placement vs CGPA")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

```

### Dataset Overview
 The dataset contains student academic and skill-related information used to analyze
 placement outcomes. It includes numerical and categorical features such as CGPA,
 IQ score, internships, and placement status.

### Data Cleaning
 - Missing values were handled using mean imputation for numerical columns.
 - Categorical Yes/No values were converted into numerical format (1/0).
 - Column names were cleaned for consistency.

### Exploratory Data Analysis
 - Distribution plots showed that CGPA follows a near-normal distribution.
 - Box plots indicated that students with higher CGPA have higher placement chances.
 - Correlation heatmap revealed that CGPA and skill-related features have strong
  correlation with placement status.

### Key Insights
 - CGPA is one of the most important factors affecting placement.
 - Academic performance and skill indicators significantly influence placement outcomes.
 - Cleaned and processed data is now ready for machine learning models.
